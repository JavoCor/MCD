#### Haz una función que dada una matriz cualquiera, normalice cada fila restando por la media y diviendo por el desvío. Pruebala

````python
v9=np.arange(25).reshape(5,5)
````

````python
np.median(v9, axis=1)
````

````python
v9
````

````python
v9=np.arange(25).reshape(5,5)
normalized_matrix = np.zeros(25).reshape(5,5)
for i in range(5):
    median=np.median(v9[i])    
    std_dev=np.std(v9[i])
    normalized_row=(v9[i]-median)/std_dev
    normalized_matrix[i]=normalized_row
normalized_matrix
````

````python
def normalize_rows(matrix):
    row_means = np.mean(matrix, axis=1)
    row_stds = np.std(matrix, axis=1)
    normalized_matrix = (matrix - row_means[:, np.newaxis]) / row_stds[:, np.newaxis]
    return normalized_matrix
normalize_rows(v9)
````

#### Haz una función que dada una matriz cualquiera, normalice cada columna restando por la media y diviendo por el desvío. Pruebala

````python
v9=np.arange(25).reshape(5,5)
normalized_matrix = np.zeros(25).reshape(5,5)
for i in range(5):
    median=np.median(v9[:,i])    
    std_dev=np.std(v9[:,i])
    normalized_row=(v9[:,i]-median)/std_dev
    normalized_matrix[:,i]=normalized_row
normalized_matrix
````
